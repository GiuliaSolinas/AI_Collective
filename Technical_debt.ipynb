```python
import matplotlib.pyplot as plt
import numpy as np

# Time (in months)
time = np.arange(0, 24, 1)

# Cost of maintaining clean code (steady, low slope)
clean_code_cost = 50 + time * 2

# Cost of maintaining code with technical debt (starts low but increases steeply over time)
tech_debt_cost = 50 + time * 2 + np.exp(time / 8)

plt.figure(figsize=(10, 6))
plt.plot(time, clean_code_cost, label='Integrated Solution Maintenance Cost', color='green')
plt.plot(time, tech_debt_cost, label='Solution w. Technical Debt Maintenance Cost', color='red')
plt.xlabel('Time (months)')
plt.ylabel('Cost')
plt.title('Technical Debt - Mind the gap')
plt.legend()
plt.grid(True)
plt.show()
```
